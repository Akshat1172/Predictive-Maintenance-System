# Setup and Imports


```python
from pathlib import Path
import sys

import pandas as pd
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    f1_score,
    precision_score,
    recall_score,
)
from sklearn.model_selection import train_test_split

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT))

from src.data.load_data import load_data
from src.models.train import build_decision_tree_pipeline
from src.models.train import build_logistic_pipeline
from src.models.train import build_random_forest_pipeline
from src.models.train import build_xgboost_pipeline
```

# Load Data


```python
df_raw, df = load_data()
```

# Prepare Features and Target


```python
columns_to_drop = [
    "UDI",
    "Product ID",
    "TWF",
    "HDF",
    "PWF",
    "OSF",
    "RNF",
]

df = df.drop(columns=columns_to_drop)

X = df.drop(columns=["Machine failure"])
y = df["Machine failure"]
```

# Train/Test Split


```python
TEST_SIZE = 0.2
RANDOM_STATE = 42

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y,
)
```

# Logistic Regression


```python
logistic_pipeline = build_logistic_pipeline()
```

```python
logistic_pipeline.fit(X_train, y_train)
```

```python
logistic_predictions = logistic_pipeline.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, logistic_predictions))

print(classification_report(y_test, logistic_predictions))

print(confusion_matrix(y_test, logistic_predictions))
```

# Decision Tree


```python
dt_pipeline = build_decision_tree_pipeline()
```

```python
dt_pipeline.fit(X_train, y_train)
```

```python
dt_predictions = dt_pipeline.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, dt_predictions))

print(classification_report(y_test, dt_predictions))

print(confusion_matrix(y_test, dt_predictions))
```

```python
dt_train_predictions = dt_pipeline.predict(X_train)
dt_test_predictions = dt_pipeline.predict(X_test)

print("Training Accuracy:", accuracy_score(y_train, dt_train_predictions))
print("Testing Accuracy :", accuracy_score(y_test, dt_test_predictions))
```

# Random Forest


```python
rf_pipeline = build_random_forest_pipeline()
rf_pipeline.fit(X_train, y_train)

rf_predictions = rf_pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, rf_predictions))
print(classification_report(y_test, rf_predictions))
print(confusion_matrix(y_test, rf_predictions))
```

```python
rf_train_predictions = rf_pipeline.predict(X_train)

print("Training Accuracy:", accuracy_score(y_train, rf_train_predictions))
print("Testing Accuracy :", accuracy_score(y_test, rf_predictions))
```

# XGBoost Baseline


```python
xgb_pipeline = build_xgboost_pipeline()

xgb_pipeline.fit(X_train, y_train)
```

```python
xgb_predictions = xgb_pipeline.predict(X_test)

print(
    "Accuracy:",
    accuracy_score(y_test, xgb_predictions),
)

print(
    classification_report(y_test, xgb_predictions)
)

print(
    confusion_matrix(y_test, xgb_predictions)
)
```

```python
xgb_train_predictions = xgb_pipeline.predict(X_train)

print(
    "Training Accuracy:",
    accuracy_score(y_train, xgb_train_predictions),
)

print(
    "Testing Accuracy:",
    accuracy_score(y_test, xgb_predictions),
)
```
